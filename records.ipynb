```python
import datetime as dt
import pandas as pd
from ggplot import *
%matplotlib inline
```

```python
data = {
    'Time Date': [dt.datetime(2015,6,24,2,55),
                  dt.datetime(2015,6,24,22,15),
                  dt.datetime(2015,6,26,0,38),
                  dt.datetime(2015,6,27,1,48),
                  dt.datetime(2015,6,27,23,5),
                  dt.datetime(2015,6,29,0,2),
                  dt.datetime(2015,6,30,1,41),
                  dt.datetime(2015,7,1,0,5),
                  dt.datetime(2015,7,1,21,48),
                  dt.datetime(2015,7,3,1,48)],
    'SRR': [33,
            25,
            21,
            32,
            31,
            26,
            28,
            23,
            25,
            24],
    'HR':  [88,
            120,
            100,
            92,
            108,
            104,
            84,
            96,
            108,
            104],
    'Notes': ['Skips beat ~8x/min',
              'Much less lethargic. HR while wide awake and now using stethoscope',
              'HR while resting. SRR irregular',
              'More lethargic. Poor appetite. HR bit irregular',
              'Marginally less lethargic. HR bit irregular',
              'Less lethargic',
              'Improved appetite. Did heavier excercise, but quickly exhausted by it.',
              'Same',
              'Marginally more lethargic',
              'Much more lethargic and worse appetite, particularly in the morning']
}

df = pd.DataFrame(data)
df
```

```python
print ggplot(df, aes(x='Time Date', y='SRR')) +\
    ggtitle('Sleeping Respiratory Rate against Time and Date') +\
    xlab('Time and Date') +\
    scale_y_continuous(name='SSR (Breaths/min)', limits=(0,40)) +\
    geom_line(aes(y='SRR'), color='blue') +\
    geom_point(aes(y='SRR'), color='blue') +\
    scale_x_date(labels=date_format('%a %b-%d'), breaks='1 day')
```

```python
print ggplot(df, aes(x='Time Date', y='HR')) +\
    ggtitle('Heart Rate against Time and Date') +\
    xlab('Time and Date') +\
    scale_y_continuous(name='HR (Beats/min)', limits=(0,150)) +\
    geom_line(aes(y='HR'), color='red') +\
    geom_point(aes(y='HR'), color='red') +\
    scale_x_date(labels=date_format('%a %b-%d'), breaks='1 day')
```
