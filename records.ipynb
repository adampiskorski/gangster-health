```python
import datetime as dt
import pandas as pd
from ggplot import *
%matplotlib inline
```

```python
data = {
    'Time Date': [dt.datetime(2015,6,24,2,55),
                  dt.datetime(2015,6,24,22,15),
                  dt.datetime(2015,6,26,0,38),
                  dt.datetime(2015,6,27,1,48)],
    'SRR': [33,
            25,
            21,
            32],
    'HR':  [88,
            120,
            100,
            92],
    'Notes': ['Skips beat ~8x/min',
              'Much less lethargic. HR while wide awake and now using stethoscope',
              'HR while resting. SRR irregular',
              'More lethargic. Poor appetite. HR bit irregular']
}

df = pd.DataFrame(data)
df
```

```python
print ggplot(df, aes(x='Time Date')) +\
    ggtitle('Sleeping Respiratory Rate against Time and Date') +\
    xlab('Time and Date') +\
    ylab('SSR (Breaths/min)') +\
    geom_line(aes(y="SRR"), color="blue") +\
    geom_point(aes(y="SRR"), color="blue")
```

```python
print ggplot(df, aes(x='Time Date')) +\
    ggtitle('Heart Rate against Time and Date') +\
    xlab('Time and Date') +\
    ylab('HR (Beats/min)') +\
    geom_line(aes(y="HR"), color="red") +\
    geom_point(aes(y="HR"), color="red")
```
