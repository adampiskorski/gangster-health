```python
import datetime as dt
import pandas as pd
from ggplot import *
%matplotlib inline
```

```python
data = {
    'Time': [dt.datetime(2015,6,23,20,13), 
             dt.datetime(2015,6,24,19,5),
             dt.datetime(2015,6,25,21,50)],
    'SRR': [30,
            31,
            29],
    'HR':  [149,
            150,
            151]
}

df = pd.DataFrame(data)
df
```

```python
print ggplot(df, aes(x='Time')) +\
    ggtitle('Sleeping Respiratory Rate against Time and Date') +\
    xlab('Time and Date') +\
    ylab('SSR (Breaths/min)') +\
    geom_line(aes(y="SRR"), color="blue") +\
    geom_point(aes(y="SRR"), color="blue")
```

```python
print ggplot(df, aes(x='Time')) +\
    ggtitle('Heart Rate against Time and Date') +\
    xlab('Time and Date') +\
    ylab('HR (Beats/min)') +\
    geom_line(aes(y="HR"), color="red") +\
    geom_point(aes(y="HR"), color="red")
```
